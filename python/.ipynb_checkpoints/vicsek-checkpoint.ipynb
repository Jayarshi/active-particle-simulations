```python
import numpy as np
import math as mt
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import scipy as sc
import warnings
import time
from numba import njit
#import myuncflib
```

```python
@njit
def getY(x,theta,boxL):
    cos = np.cos(theta)
    sin = np.sin(theta)
    m = boxL/25
    trig = m*np.column_stack((cos,sin))
    y1 = x[:,0]+trig[:,0]
    y2 = x[:,1]+trig[:,1]
    y = np.column_stack((y1,y2))
    return y
```

```python
num_part = 1000
np.random.seed(100)
boxL = 5

x = np.random.random((num_part,2))*boxL
#theta = np.random.random((num_part,1))*np.pi*2
theta = np.random.random((num_part,1))*2*np.pi
#Parameters
boxL = 5

y =  getY(x,theta,boxL)

#Plotting
dx = y[:,0] - x[:,0]
dy = y[:,1] - x[:,1]
plt.quiver(x[:,0], x[:,1], dx, dy,
           angles='xy',
           scale_units='xy',
           scale=2,
           color='red')
```

```python
@njit
def config_avg(x,theta,r):
    avgt = np.zeros_like(theta)

    for i in range(len(x)):
        sx = 0.0
        sy = 0.0
        for j in range(len(x)):
            dx = x[i,0]-x[j,0]
            dy = x[i,1]-x[j,1]
            dist = np.sqrt(dx*dx + dy*dy)
            if dist <= r:
                sx += np.cos(theta[j,0])
                sy += np.sin(theta[j,0])
        avgt[i,0] = np.arctan2(sy,sx)
    return avgt
```

```python
@njit
def vel_vec(v,theta):
    cos = np.cos(theta)
    sin = np.sin(theta)
    vec = v*np.column_stack((cos,sin))
    return vec
```

```python
@njit
def evolve(x,theta,n,rad_int,noise,v,dt,boxL):
    num_part = len(x)
    for step in range(n):
        s = np.random.random((num_part,1)) - 0.5
        avgt = config_avg(x,theta,rad_int)
        theta = avgt + noise*s*dt
        cos = np.cos(theta)
        sin = np.sin(theta)
        x[:,0] += v*cos[:,0]*dt
        x[:,1] += v*sin[:,0]*dt
        x[:,0] = x[:,0] % boxL
        x[:,1] = x[:,1] % boxL
    return x,theta
```

```python
#Initialise Distribution
num_part = 1000
np.random.seed(100)
x = np.random.random((num_part,2))*boxL
#theta = np.random.random((num_part,1))*np.pi*2
theta = (np.random.randint(2, size=num_part)*np.pi).reshape(num_part,1)
#Parameters
boxL = 5
noise = 0.1 #strength of noise
v = 0.5  #Constant Velocity Term
rad_int = 0.1 #Interaction Radius


#Update
t = 500 #time of evolution
dt = 0.01
#sqdt=np.sqrt(dt)
n = int(t/dt)
x,theta = evolve(x,theta,n,rad_int,noise,v,dt,boxL)
y =  getY(x,theta,boxL)

#Plotting
dx = y[:,0] - x[:,0]
dy = y[:,1] - x[:,1]
plt.quiver(x[:,0], x[:,1], dx, dy,
           angles='xy',
           scale_units='xy',
           scale=2.5,
           color='red')
```
